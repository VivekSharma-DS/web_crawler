```python
from bs4 import BeautifulSoup
import requests

url = "https://pureportal.coventry.ac.uk/en/organisations/school-of-life-sciences/publications/"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
all_publication = soup.find_all("div", class_="rendering rendering_researchoutput rendering_researchoutput_short rendering_contributiontojournal rendering_short rendering_contributiontojournal_short")
#print(all_publication)
for person in all_publication:
    author = person.find("a", class_="link person")
    publication = person.find("h3", class_="title")
    publication_link = person.find("a", class_="link")
    publication_date = person.find("span", class_="date")
    print(author.string)
    print(publication.string)
    print(publication_date.string)

    print(publication_link.get('href'))
    print(author['href'])
    print("----------------------------------------------\n")
```

```python
from bs4 import BeautifulSoup
import requests

info = {}
url = "https://pureportal.coventry.ac.uk/en/publications/a-comparison-of-physical-activity-between-home-based-and-centre-b"
page = requests.get(url)
soup = BeautifulSoup(page.text, "html.parser")
all_author = soup.find("p", class_="relations persons")
print(all_author)
#for sub_author in all_author:
#    print(sub_author.string)

authors = soup.find("p", class_="relations persons").text.split(",")
authors = [author.strip() for author in authors]
authors = dict.fromkeys(authors, None)

for authorLinks in soup.find("p", class_="relations persons").find_all("a", class_="link person", href=True):
    authors[authorLinks.text.strip()] = authorLinks["href"]

    
    info["authors"] = authors

print(info)
```
